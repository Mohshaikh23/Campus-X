```python
death = pd.read_csv('time_series_covid19_deaths_global.csv')
confirmed= pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
death.head()
```

```python
death.shape
```

```python
confirmed.head()
```

```python
confirmed.shape
```

```python
confirmed_long = confirmed.melt(id_vars = ['Province/State','Country/Region','Lat','Long'], var_name='Date',value_name='Confirmed')
```

```python
confirmed_long.shape
```

```python
death_long = death.melt(id_vars = ['Province/State','Country/Region','Lat','Long'], var_name='Date',value_name='Deaths')
```

```python
death_long.shape
```

```python
ndf = confirmed_long.merge(death_long,on=['Province/State','Country/Region','Lat','Long','Date'])
ndf
```
