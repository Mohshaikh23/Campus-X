```python
import numpy as np
import pandas as pd
```

```python
ipl_matches = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv"
```

```python
df= pd.read_csv(ipl_matches)
```

```python
df.head()
```

```python
list(set(list(df["Team1"]) + list(df["Team2"])))
```

```python
def team_data(team1, team2):
    mask = df[(df["Team1"] == team1) & (df["Team2"]== team2) | (df["Team1"]== team2) & (df["Team2"]== team1)]
    
    matches_won_team1 = mask["WinningTeam"].value_counts()[team1]
    matches_won_team2 = mask["WinningTeam"].value_counts()[team2]
    draw_matches = mask["WinningTeam"].value_counts().shape[0] - (matches_won_team1 + matches_won_team2)

    response = {"Team1":team1,
                "Team2":team2,
                "Matches Won by Team1":matches_won_team1,
                "Matches Won by Team2":matches_won_team2,
                "Matches Drawn":draw_matches,
               }
    
    return response
team_data("Rajasthan Royals","Royal Challengers Bangalore")
```
